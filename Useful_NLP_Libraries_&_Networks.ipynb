Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.

Answer:

Natural Language Toolkit (NLTK) and spaCy are two popular libraries used for Natural Language Processing in Python. Both are widely used but serve slightly different purposes.
Features

NLTK is mainly designed for education and research. It provides a large number of NLP algorithms, corpora, and tools for tasks such as tokenization, stemming, tagging, parsing, and classification. It is very useful for learning NLP concepts.

spaCy, on the other hand, is designed for production-level NLP applications. It provides efficient implementations for tasks such as tokenization, part-of-speech tagging, named entity recognition, dependency parsing, and word vectors.

Ease of Use

NLTK requires more manual setup and sometimes multiple steps to perform tasks. It is flexible but may require more coding.
spaCy is easier to use for many tasks because it provides pre-trained pipelines that perform several operations automatically.

Performance

NLTK is slower compared to spaCy because it focuses more on teaching and experimentation.
spaCy is optimized for speed and uses Cython-based implementations, making it significantly faster for large-scale NLP applications.
In summary, NLTK is better for learning and experimentation, while spaCy is better for real-world industrial NLP applications.

Question 4: Describe the architecture and functioning of a Recurrent Neural Network (RNN).

Answer:

A Recurrent Neural Network (RNN) is a type of neural network designed to process sequential data such as text, speech, or time series.
Unlike traditional neural networks, RNNs have connections that form cycles, allowing them to retain information from previous inputs.

Architecture

An RNN consists of three main components:

Input layer

Hidden layer (recurrent connections)

Output layer

The key feature of an RNN is the hidden state, which acts as memory and carries information from one time step to the next.

Functioning

When processing text, an RNN reads words sequentially. At each step:
The current word is given as input.

The network combines the current input with the previous hidden state.

A new hidden state is produced.

The network generates an output prediction.

This sequential processing allows the model to understand context and word order in sentences.

RNNs are commonly used in:

Language modeling

Machine translation

Speech recognition

Text generation

However, standard RNNs suffer from problems such as vanishing gradients,
which led to the development of improved architectures like LSTM and GRU.

Question 5: What is the key difference between LSTM and GRU networks in NLP applications?

Answer:

Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) are advanced versions of Recurrent Neural Networks designed to solve the vanishing gradient problem.

LSTM

LSTM networks have a complex architecture with three main gates:

Input gate

Forget gate

Output gate

These gates control how information is stored, updated, and removed from memory.

LSTMs are powerful for learning long-term dependencies in text sequences.

GRU

GRU is a simplified version of LSTM. It combines some gates together and uses only two gates:

Update gate

Reset gate

Because of its simpler structure, GRU models are faster to train.
Key Difference

The main difference is that LSTM has more gates and a separate memory cell, while GRU has fewer parameters and simpler architecture.

As a result:

LSTM → more complex but powerful

GRU → faster and computationally efficient

Both are widely used in NLP tasks like text classification and language modeling.

```python
#Question 7: Tokenization and Frequency Distribution using NLTK

import nltk
nltk.download('punkt_tab') # Correcting the download to 'punkt_tab' as indicated by the error
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language."""

tokens = word_tokenize(text.lower())

freq = FreqDist(tokens)

print("Tokens:")
print(tokens)

print("\nTop Words:")
print(freq.most_common(10))
```

```python
#Question 8: Basic LSTM model using Keras

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
"I love this project",
"This is an amazing experience",
"I hate waiting in line",
"This is the worst service",
"Absolutely fantastic"
]

labels = [1,1,0,0,1]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

padded = pad_sequences(sequences, maxlen=5)

# Calculate vocabulary size for the Embedding layer (max word index + 1 for padding)
vocab_size = len(tokenizer.word_index) + 1

model = Sequential()
# Removed 'input_length' as it is deprecated and updated 'input_dim'
model.add(Embedding(input_dim=vocab_size, output_dim=16))
model.add(LSTM(16))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Convert padded sequences and labels to TensorFlow tensors
padded_tensor = tf.constant(padded, dtype=tf.int32)
labels_tensor = tf.constant(labels, dtype=tf.float32)

model.fit(padded_tensor, labels_tensor, epochs=10)

print("Model trained successfully")
```

```python
#Question 9: spaCy NLP pipeline (Tokenization, Lemmatization, Entity Recognition)

import spacy

nlp = spacy.load("en_core_web_sm")

text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role
in the development of India’s atomic energy program."""

doc = nlp(text)

print("Tokens and Lemmas:")
for token in doc:
    print(token.text, "->", token.lemma_)

print("\nNamed Entities:")
for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```
